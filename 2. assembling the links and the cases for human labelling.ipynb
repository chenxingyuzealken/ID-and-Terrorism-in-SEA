```python
import pandas as pd
from IPython.display import display
```

```python
self_source_df = pd.read_csv(r'C:\Users\User\Desktop\00_Pipeline_0\0.1 Data Collection using self-sourced data\self_sourced_ner_url_list.csv',header=None)
gtd_df = pd.read_csv(r'C:\Users\User\Desktop\00_Pipeline_0\0.2 Data Collection using GTD\ML_GTD_Links_list.csv')
```

```python
self_source_df.columns = ['index','link','ID']
self_source_df.head().append(self_source_df.tail())
```

```python
gtd_df = gtd_df.drop(['Unnamed: 0','dd'],axis=1,)
gtd_df.head().append(gtd_df.tail())
```

```python
#combining all the predicted CSVs together

def create_score_df(file,predict):
    df = pd.read_csv(file,header=None)
    df.columns = ['junk','score','txt','link']
    df['predicted_category'] = predict
    df = df.drop(['junk','txt'],axis=1)
    #removing those of a certain threshold and below
    df = df[df['score']>0.85]
    df = df.drop_duplicates()
    df.sort_values(by='score',ascending=False)
    return df

ct_df = create_score_df('highest_link_score_ConspiracyTheory.csv','conspiracy theory')
ru_df = create_score_df('highest_link_score_rumour.csv','rumour')
sm_df = create_score_df('highest_link_score_smear.csv','smear')
id_df = create_score_df('highest_link_score_identity.csv','identity')
fe_df = create_score_df('highest_link_score_fear.csv','fear')
sa_df = create_score_df('highest_link_score_safety.csv','safety')
```

```python
df = pd.concat([ct_df,ru_df,sm_df,id_df,fe_df,sa_df])
df = df.drop_duplicates(subset=['link'])
```

```python
gtd = pd.merge(df,gtd_df,on='link').drop_duplicates(subset=['link'])
display(gtd.sort_values(by='GTD_ID',ascending=False))
gtd['GTD_ID'].value_counts()
```

```python
self_source = pd.merge(df,self_source_df,on='link').drop_duplicates(subset=['link'])
display(self_source)
self_source['ID'].value_counts()
```

```python
gtd.to_csv('gtd.csv')
```
